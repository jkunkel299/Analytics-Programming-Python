## Question 2: String Manipulation

```python
""" Load the assigned string into the Jupyter notebook """

s = "I am happy to join with you today in what will go down in history as the greatest demonstration for freedom in the history of our nation. Five score years ago, a great American, in whose symbolic shadow we stand today, signed the Emancipation Proclamation. This momentous decree came as a great beacon light of hope to millions of Negro slaves who had been seared in the flames of withering injustice. It came as a joyous daybreak to end the long night of their captivity. But one hundred years later, the Negro still is not free. One hundred years later, the life of the Negro is still sadly crippled by the manacles of segregation and the chains of discrimination. One hundred years later, the Negro lives on a lonely island of poverty in the midst of a vast ocean of material prosperity. One hundred years later, the Negro is still languishing in the corners of American society and finds himself an exile in his own land. So we have come here today to dramatize a shameful condition."
```

### 1. Find out how many unique words in s.

To start, I chose to convert all the characters to lowercase. Looking through the readings for this week, I did not find any one function that would quickly solve this task, so I decided that splitting the string into a string array would likely be a logical step in my solution. None of the questions were concerned with punctuation or the number of sentences within the string, so I also decided to remove all the punctuation from the string. 

```python
""" prepare the string array:  remove punctuation marks, convert 
all characters to lowercase, and split the string into an array."""

# import string in order to use the predefined string.punctuation
import string

# use comprehension in the join function to remove punctuation marks
s_clean = ''.join(char for char in s if char not in string.punctuation)
# convert to lowercase
s_lower = s_clean.lower()
# create the string array by splitting using a space ' ' as the delimiter
s_arr = s_lower.split(' ')
```

In order to find the number of unique words in the string, I decided to use a for-loop. I created an empty string `unique` and for each word in the string array, if the word was not already in `unique`, appended the word to the array. or comparative purposes, I chose to print, in markdown, the number of words in the initial string array versus the number of unique words in the speech.

```python
""" find the number of unique words in the speech, using the 
string array s_arr """

unique = []
for s in s_arr:
    if s not in unique:
        unique.append(s)

len(unique)
printmd(f"The number of words in the string is {len(s_arr)}. The number of unique words in the string is {len(unique)}.")
```

### 2. Which word appears the most?

To determine which word appears the most, I decided to make a dictionary of unique words and the number of times they appear in the speech. I used a dictionary comprehension, where the key is a word in the `unique` list, and the value is the count of times the word appears in the initial `s_arr`. For demonstrative purposes, I chose to print the dictionary of unique words and the number of times they appear in the speech.

```python
""" create a dictionary of unique words and the number of times 
they appear in the speech s_arr """

unique_dict = {word: s_arr.count(word) for word in unique}
unique_dict
```

I then used dictionary functions to determine the word that appears the most in the speech and the number of occurrences.

```python
""" find the maximum value in the dictionary and the 
corresponding key """

# find the maximum value in the dict
max_count = max(unique_dict.values())
# find the word with the maximum value in the dict
max_word = max(unique_dict, key=unique_dict.get)

# print, in markdown, the word with the most uses and its number of occurrences
printmd(f"The word that appears the most in the speech is \"{max_word}\" with {max_count} occurrences.")
```

### 3. How many words start with ‘t’?

In order to count the number of words that start with the letter 't', I chose to use a for-loop. I initialized a variable `count` to count the number of words, then looped through the list of unique words in the speech, incrementing the count if `word.startswith("t")` returned true. I then printed, in markdown, the number of words in the speech that start with the letter 't'.

```python
""" use a simple for-loop and the startswith method to 
find the number of words in the speech that start with the 
letter 't' """

# initialize count
count = 0

# use startswith method to test if each word in the unique word list starts with the letter 't'
for word in unique:
    if word.startswith("t"):
        # if true, increment count by 1
        count += 1

# print, in markdown, the number of words that start with the letter 't'
printmd(f"The number of words in the speech that start with the letter \'t\' is {count}.")
```
