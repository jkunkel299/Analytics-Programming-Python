#### Question 1
1. Use the randn function to create an array with a dimension of 5X5 and use a for loop to calculate the sum of all elements in the diagonal of the array.

```python
#!/usr/bin/env python
""" Import display functions and create a function to print outputs to 
Markdown """

from IPython.display import display, Markdown
def printmd(string):
    display(Markdown(string))
```

```python
""" Use NumPy to generate a random 5x5 array. The array is printed.
The elements along the diagonal of the array are summed and printed 
as Markdown. """

import numpy as np

arr_x = np.random.randn(5, 5)

sum_arr = 0

for i in range(5):
    sum_arr += arr_x[i, i]

printmd("Random 5x5 array:")
print(f"{arr_x}\n")
printmd(f"Sum of all elements in the diagonal of the arrray = {sum_arr}")
```

2. Choose any three functions to apply to this array.

```python
""" The array is manipulated to calculate the mean of all elements in the 
array (function 1), the sum of the elements in each row (function 2), and 
to take the absolute value of negative elements in the array and return 
the array as a whole (function 3). """

printmd(f"Mean of the array: {arr_x.mean()}")
printmd(f"Sum across the rows: {arr_x.sum(0)}")
printmd(f"Changing negative elements in the array to their absolute values: \n")
print(f"{np.where(arr_x < 0, abs(arr_x), arr_x)}")
```

1. Use x = np.random.randint(0, 1000, size = (10, 10)) to generate 10x10 array and use a for loop to find out how many even numbers are in it.

```python
""" Use NumPy to generate a 10x10 array of random integers between 
0 and 1000 (np.random.randint). Then, use a for-loop to inspect each 
element. If the element is even, the count of even numbers (count_even) 
in the array is incremented by 1. The array is printed and the count of
even numbers is printed as Markdown. """

x = np.random.randint(0, 1000, size = (10, 10))
count_even = 0

for row in x:
    for element in row:
        if element % 2 == 0:
            count_even += 1

printmd("Random 10x10 array:")
print(f"{x}\n")
printmd(f"The number of even elements in the array \"x\" is {count_even}")
```

2. Randomly generate an 8x9 array from a normal distribution with mean = 1, sigma = 0.5. Calculate the mean of elements whose indexes have a relation of (i+j)%5 == 0  (i is row index and j is column index).

```python
""" Use NumPy to generate an 8x9 array from a normal distribution 
(np.random.normal) with mean = 1 (loc=1.0), sigma = 0.5 (scale=0.5).
The array is printed. Then, each element in the array is inspected 
using the relation (i + j) % 5 == 0, and elements meeting the condition
are added to the list rel_elements. The list is converted into an 
ndarray np_rel, and the mean of the elements whose indexes met the 
condition is calculated using np_rel.mean(). The elements are printed,
and the mean of the elements is displayed in Markdown. """

norm = np.random.normal(loc=1.0, scale=0.5, size=(8, 9))

printmd("8x9 array from normal distribution:")
print(f"{norm}\n")

rel_elements = []

for i in range(8):
    for j in range(9):
        if (i + j) % 5 == 0:
            rel_elements.append(norm[i,j])

np_rel = np.array(rel_elements)
mean_rel = np_rel.mean()

printmd(f"The elements whose indexes have a relation of (i+j)%5 == 0:")
print(f"{np_rel}\n")
printmd(f"The mean of the elements whose indexes have a relation of (i+j)%5 == 0 is {mean_rel}")
```
