### Question 1

```python
# Code block provided in question
import numpy as np
L1 = []
np.random.seed(56)
for i in np.random.randint(0, 100, 10):
    L1.extend([i]*np.random.randint(0, 100, 1)[0])
np.random.shuffle(L1)

L1_int = [int(x) for x in L1]

# What are the unique values?
unique_L1 = sorted(list(set(L1_int)))
print('Unique values: ', unique_L1)

# How many unique values?
num_unique_L1 = len(unique_L1)
print('Number of unique values: ', num_unique_L1)

# Create a dictionary with the unique items in L1 as keys and their count as dictionary values
unique_list = list(unique_L1)
unique_list.sort()

dict_unique = {key: 0 for key in unique_list}
for num in L1_int:
    if num in dict_unique:
        dict_unique[num] += 1


# Which value appears most frequently?
max_occ = max(dict_unique.values())
max_occ_key = [key for key in dict_unique if dict_unique[key] == max_occ]

print('The value that appears most frequently is ', max_occ_key[0], ' with ', max_occ, ' occurrences')
```

```python
# Implement the function pow(x, n), which calculates x raised to the power n (xn). Please don't use x**n. 
def pow(x, n):
    abs_n = abs(n)
    ans = 1
    for _ in range(abs_n):
        ans *= x
    if n < 0:
        ans = 1 / ans
    return ans

# Calculate pow(2, 10) and pow(3, -3)
print('pow(2, 10) = ', pow(2, 10))
print('pow(3, -3) = ', pow(3, -3))
```
